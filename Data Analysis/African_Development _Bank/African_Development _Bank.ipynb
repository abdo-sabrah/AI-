```python
import pandas as pd
import numpy as np
import seaborn as sns
import datetime as dt
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('finance1.csv',encoding='latin -1')
df.head(10)
```

```python
df.tail()
```

```python
df.info()
```

```python
df2 = df.dropna()
```

```python
df2['Starting Date']=pd.to_datetime(df2['Starting Date'])
df2['Starting Date'].dtype
```

```python
df2['Starting Year'] = df2['Starting Date'].dt.year
```

```python
df2.head()
```

```python
df3 =df2[ df2['Starting Year']==2019]
df3.head()
```

```python
df3= df3.drop(['Project Code','Title','Starting Date','Status', 'Source of Financing','Sector','Sovereign'],axis=1)
df3.head()
```

```python
df3.dtypes
```

```python
df3['Commitment in U.A'] = df3['Commitment in U.A'].str.replace(r',', '')
df3.head()
```

```python
df3['Commitment in U.A'] = df3['Commitment in U.A'].apply(pd.to_numeric)
df3['Commitment in U.A'].dtype
```

```python
max_country= df3.groupby(['Country'])['Commitment in U.A'].agg(['max'])
max_country.sort_values(by='max', ascending=False).head(1)
```

```python
min_country= df3.groupby(['Country'])['Commitment in U.A'].agg(['min'])
min_country.sort_values(by='min', ascending=False).head(1)
```

```python
df3.plot(x='Country',y='Commitment in U.A',kind = 'bar', figsize=(8,8))
```

```python
df4= df3.groupby(['Country']).sum()
df4.reset_index(inplace=True)
df4.head()
```

```python
plt.figure(figsize=(10,8))
plt.bar(df4['Country'],df4['Commitment in U.A'], label='Country')
plt.xticks(rotation=90)
plt.title("Commitment in U.A and Country")
plt.xlabel('Country')
plt.ylabel('Commitment in U.A')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=[20,10])

labels=['Botswana', 'Kenya',  'Morocco','Tunisia','Mozambique', 'Multinational', 'Nigeria',
        'Rwanda', 'Senegal','Mali', 'Guinea','Madagascar']

plt.pie(df4['Commitment in U.A'],labels=labels,autopct='%.2f')
```
